## 4. Removing Punctuation

```python
import string
puctuations = string.punctuation
```

```python
def remove_puntuations(sent):
    return sent.translate(str.maketrans("","",puctuations))
```

```python
df['review'] = df['review'].apply(remove_puntuations)
```

## 5. Chat Word Treatment (Slang Expansion)

eg: ASAP -> As Soon As Possible

```python
chat_words_dict = {}
with open("data/english_chat_words.txt","r") as file:
    for f in file.readlines():
        chat_words_dict[f.split("->")[0].strip()] = f.split("->")[1].strip()
```

```python
def remove_chatwords(sentence):
    result = str(sentence)
    for word in sentence.split(" "):
        if word.upper() in chat_words_dict:
            print(word)
            print(chat_words_dict[word.upper()])
            result = result.replace(word, chat_words_dict[word.upper()])
    return result
```

```python
df['review'] = df['review'].apply(remove_chatwords)
```

## Removing stopwords

```python
from nltk.corpus import stopwords
import nltk
nltk.data.path.append('/home/rathan-r-nayak/Rathan/Machine Learning/machine-leaning/NLP/TextPreprocessing/data')
stopwords.words('english')
```

```python
def remove_stopwords(sentence):
    li = []

    for word in sentence.split(" "):
        if word not in stopwords.words('english'):
            li.append(word)
    return " ".join(li)
```

```python
df['review'] = df['review'].apply(remove_stopwords)
```

```python
from symspellpy import SymSpell, Verbosity
import pkg_resources

symspell = SymSpell(max_dictionary_edit_distance=2)

dictionary_path = pkg_resources.resource_filename(
    "symspellpy", "frequency_dictionary_en_82_765.txt",
)
symspell.load_dictionary(dictionary_path, term_index=0, count_index=1)


def fix_spelling(sentence):
    suggestions = symspell.lookup_compound(sentence, max_edit_distance=2)
    return suggestions[0].term
```

```python
df['review'] = df['review'].apply(fix_spelling)
```

## Lemmatizer

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
```

```python
nltk.pos_tag(nltk.word_tokenize("one reviewers mention watch 1 oz episode youll"))
```

```python
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
lemmatizer = WordNetLemmatizer()

def lemmatize_words(sentence):
    tokens = nltk.word_tokenize(sentence)
    pos_tags = nltk.pos_tag(tokens=tokens)
    result = []
    
    for word in pos_tags:
        result.append(lemmatizer.lemmatize(word=word[0],pos=get_wordnet_pos(word[1])))
    return " ".join(result)
```

```python
df['review'] = df['review'].apply(lemmatize_words)
```
