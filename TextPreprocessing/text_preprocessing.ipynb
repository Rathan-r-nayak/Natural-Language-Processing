## 4. Removing Punctuation

### Using for loop

```python
import string
```

```python
puctuations = string.punctuation
puctuations
```

```python
# will take more time

def remove_puctuations(word):
    for c in puctuations:
        word = word.replace(c,"")
    return word
```

### Using translate

```python
# take less time

def remove_puctuations2(word):
    return word.translate(str.maketrans("","",puctuations))
```

## 5. Chat Word Treatment (Slang Expansion)

eg: ASAP -> As Soon As Possible

```python
chat_words_dict = {}
with open("data/english_chat_words.txt","r") as file:
    for f in file.readlines():
        chat_words_dict[f.split("->")[0].strip()] = f.split("->")[1].strip()

    # chat_words.append(file.read().split("->"))
```

```python
def remove_chatwords(sentence):
    result = str(sentence)
    for word in sentence.split(" "):
        if word.upper() in chat_words_dict:
            print(word)
            print(chat_words_dict[word.upper()])
            result = result.replace(word, chat_words_dict[word.upper()])
    return result
```

```python
remove_chatwords("TBH I have to go ASAP to office")
```

```python
df['description'].apply(remove_chatwords)
```

```python
df['description'] = df['description'].apply(remove_chatwords)
```

## Removing stopwords

```python
import nltk
nltk.download('stopwords', download_dir='/home/rathan-r-nayak/Rathan/Machine Learning/machine-leaning/NLP/TextPreprocessing/data')
from nltk.corpus import stopwords
nltk.data.path.append(
    '/home/rathan-r-nayak/Rathan/Machine Learning/machine-leaning/NLP/TextPreprocessing/data'
)
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(sentence):
    li = []

    for word in sentence.split(" "):
        if word not in stopwords.words('english'):
            li.append(word)
    return " ".join(li)
```

```python
remove_stopwords("I was in office for long time")
```

```python
df['description'].apply(remove_stopwords)
```

```python
df['description'] = df['description'].apply(remove_stopwords)
```
