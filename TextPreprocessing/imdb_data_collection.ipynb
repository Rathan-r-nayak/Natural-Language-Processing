```python
import pandas as pd
import requests
import os
```

```python
end_point = "https://api.themoviedb.org/3"
api_key = os.getenv("TMDB_API_KEY")
```

## Get Genre

```python
genres_response = requests.get(f"https://api.themoviedb.org/3/genre/movie/list?api_key={api_key}&language=en-US").json()

genre_dict = {}
for genre in genres_response['genres']:
    genre_dict[genre["id"]] = genre["name"]
genre_dict
```

## create dataframe and save the data

```python
columns = ["title", "genre", "language", "Actors", "description", "release_date"]

df = pd.DataFrame()
```

```python
n = 531

for i in range(1,n):
    response = requests.get(
        f"{end_point}/movie/popular",
        params = {
            "language" : "en_US",
            "api_key" : api_key,
            "page" : i
        }
    ).json()
    
    for result in response['results']:
        genres = []
        for genre_id in result["genre_ids"]:
            genres.append(genre_dict[genre_id])

        movie_id = result["id"]
        movie_casts = requests.get(f"https://api.themoviedb.org/3/movie/{movie_id}/credits?api_key={api_key}").json()
        casts = []
        for cast in movie_casts["cast"]:
            cast_name = cast["name"]
            cast_character = cast["character"]
            casts.append(f"{cast_name } ({cast_character})")
        
        row = {"title": result["title"], "genre": ", ".join(genres),"language": result["original_language"], "casts": ", ".join(casts), "description": result["overview"], "release_date": result["release_date"]}

        df = pd.concat([df, pd.DataFrame([row])], ignore_index=True)
```
